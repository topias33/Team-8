```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

scale = 1./255
train_datagen = ImageDataGenerator(rescale=scale, validation_split=0.3)
test_datagen = ImageDataGenerator(rescale=scale)
```

```python
target_size = (150,150)
batch_size = 16
class_mode = 'binary'

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size,
    batch_size=batch_size,
    class_mode=class_mode)

dev_generator = train_datagen.flow_from_directory(
    valid_dir, target_size,
    batch_size=batch_size,
    class_mode=class_mode)
```

```python
train_generator = train_datagen.flow_from_directory(
    test_dir, target_size,
    batch_size=batch_size,
    class_mode=class_mode)
```

```python
data_batch = train_generator[0][0]
imshow(data_batch[0])
show()
data_batch[0].shape
```

```python
#Build and compile the model

from tensorflow.keras import models, layers

network = models.Sequential()
network.add(layers.Conv2D(32, (3,3), activation = 'relu', padding = "same",input_shape=data_batch[0].shape))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Conv2D(32, (3,3), activation = 'relu'))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Conv2D(32, (3,3), padding = "same", activation = 'relu'))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Flatten())
network.add(layers.Dense(1, activation = 'sigmoid'))

network.summary() 
```

```python
from tensorflow.keras import optimizers

network.compile(loss='binary_crossentropy', metrics=['acc'])
```

```python
%%time
history = network.fit_generator(train_generator,
                                workers=4,
                                steps_per_epoch=10,
                                verbose=1,
                                epochs=5,
                                validation_data=dev_generator,
                                validation_steps=10)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))
```

```python
plot(epochs,acc,'bo-',label='Training acc')
plot(epochs,val_acc,'r*-',label='Validation acc')
title('Training and validation accuracy')
grid()
legend()

figure()
plot(epochs,loss,'bo-',label='Training loss')
plot(epochs,val_loss,'r*-',label='Validation loss')
title('Training and validation loss')
grid()
legend()

show()
```
